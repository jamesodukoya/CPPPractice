```python
def smallest_positive(in_list):
    smallest = float('inf')
    for num in in_list:
        if num > 0 and num < smallest:
            smallest = num
    return smallest if smallest != float('inf') else None

# Test cases
print(smallest_positive([4, -6, 7, 2, -4, 10]))
# Correct output: 2
print(smallest_positive([.2, 5, 3, -.1, 7, 7, 6]))
# Correct output: 0.2
print(smallest_positive([-6, -9, -7]))
# Correct output: None
print(smallest_positive([]))
```

```python
## Definition of the generator to produce even numbers.
def all_even():
    n = 0
    while True:
        yield n
        n += 2

my_gen = all_even()

## Generate the first 5 even numbers.
for i in range(5):
    print(next(my_gen))

## Now go and do some other processing.
do_something = 4
do_something += 3
print(do_something)

## Now go back to generating more even numbers.
for i in range(100):
    print(next(my_gen))
```

```python
def prod(a,b):
    output = a*b
    return output

def fact_gen():
    i=1
    n=i
    while True:
        output = prod(n,i)
        yield output
        i, n = i+1, output

my_gen = fact_gen()
num = 5
for i in range (num):
    print(next(my_gen))
```

```python
from collections import defaultdict as dict

def check_sudoku(square):
    num_set = set(i for i in range(1, len(square[0])+1))
    row_set = set()
    col_dict = dict(list)
    for row in square:
        for i, num in enumerate(row):
            if num in num_set:
                if num in row_set:
                    return False
                else:
                    row_set.add(num)
                if num in col_dict[i]:
                    return False
                else:
                    col_dict[i].append(num)
            else:
                return False
        row_set = set()
    return True

# Test cases
correct = [[1,2,3],
           [2,3,1],
           [3,1,2]]
incorrect = [[1,2,3,4],
             [2,3,1,3],
             [3,1,2,3],
             [4,4,4,4]]
incorrect2 = [[1,2,3,4],
             [2,3,1,4],
             [4,1,2,3],
             [3,4,1,2]]
incorrect3 = [[1,2,3,4,5],
              [2,3,1,5,6],
              [4,5,2,1,3],
              [3,4,5,2,1],
              [5,6,4,3,2]]
incorrect4 = [['a','b','c'],
              ['b','c','a'],
              ['c','a','b']]
incorrect5 = [ [1, 1.5],
               [1.5, 1]]

print(check_sudoku(incorrect))
#>>> False
print(check_sudoku(correct))
#>>> True
print(check_sudoku(incorrect2))
#>>> False
print(check_sudoku(incorrect3))
#>>> False
print(check_sudoku(incorrect4))
#>>> False
print(check_sudoku(incorrect5))
#>>> False
```
